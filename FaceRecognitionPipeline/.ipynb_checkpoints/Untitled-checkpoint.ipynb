```python
import tensorflow as tf
import numpy as np
import dlib
import imutils
import os
import cv2
from TinyFacesDetector import TinyFacesDetector
```

```python
def shape_to_np(shape, dtype="int"):
	# initialize the list of (x, y)-coordinates
    coords = np.zeros((5, 2), dtype=dtype)
	# loop over the 68 facial landmarks and convert them
	# to a 2-tuple of (x, y)-coordinates
    for i in range(0, 5):
        coords[i] = (shape.part(i).x, shape.part(i).y)
	# return the list of (x, y)-coordinates
    return coords
```

```python
def get_landmarks(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
    d_rects = tiny_faces_detector.detect(gray,nms_thresh=0.1,prob_thresh=0.5,min_conf=0.9)
    print(d_rects)
    for d in d_rects:
        x,y,w,h = d[0],d[1],d[2],d[3]
    d_rects=[dlib.rectangle(x, y, w, h)]
    #print(d_rects_tiny)
    predictor = dlib.shape_predictor('shape_predictor_5_face_landmarks.dat')
    for ret in d_rects:
        shape = predictor(gray, box=ret)
        shape=shape_to_np(shape)
        for (x, y) in shape: 
            cv2.circle(image, (x, y), 1, (0, 0, 255), -1) 
        half_face=image[ shape[2][1]-30:shape[4][1]-30, shape[2][0]-15:shape[0][0]+10 ]
    return half_face, image
```

```python
model_pkl="weights.pkl"
tiny_faces_detector = TinyFacesDetector(model_pkl,use_gpu=True)
```

```python
images = cv2.imread('/Users/krishrana/AjnaLens/Resources/Models/FaceDetectionPipeline/output/Krish/krish122.jpg') 
face, img=get_landmarks(images)

import matplotlib.pyplot as plt
plt.imshow(face)
plt.show()
```
